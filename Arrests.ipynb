```python
import folium
import requests
import pandas as pd

# For this project, I will be using the dataset from the dataMontgomery website regarding the pedestrian violations
arrest = pd.read_csv("Pedestrian_Violations_ONLY_-_Traffic_Violations_Related_to_Pedestrian_Safety.csv")
```

An explanation of your chosen dataset

I mainly chose this dataset because I want to see how frequent pedestrian violations are reported around Montgomery County. Also I wanted to know how many types of arrests are there and what kind of violations that a pedestrian could do and how frequent these arrests happened and where specifically within the Montgomery County.

```python
# We can see that this dataset has only 9159 rows and now we will see how many rows 
# That are empty/no location given by seeing the non-empty rows at the location
arrest[pd.notnull(arrest["Location"])]["Location"].count()
```

```python
# As we can see, there are no rows that has not have any location
# Now, we would create the map around the Montgomery County using the latitude and longitude of the place
map_osm = folium.Map(location=[39.09, -77.15], zoom_start=13)
map_osm
```

```python
# Seeing how many columns we have and that we can use to create for our next part
arrest.columns
```

```python
# I decided on wanting to see the types of arrest that this dataset has
# And also I am just curious on how many arrest types that are there in this dataset
arrest["Arrest Type"].unique()
```

```python
# Seeing how many in the arrest type columns that is empty
# Or if there is any NaN
arrest["Arrest Type"].count()
arrest[pd.notnull(arrest["Arrest Type"])]["Arrest Type"].count()
```

```python
# As we can see, everything is in the row and now we would decide to see the location that these violations occured
# For this instance, I would want to see the Motorcyle and Foot Patrol arrests
for index, crash in arrest.iterrows():
    if crash["Arrest Type"] == "L - Motorcycle":
        folium.Marker(location=[crash["Latitude"], crash["Longitude"]],
                    icon=folium.Icon(color='red')).add_to(map_osm)
        
    if crash["Arrest Type"] == 'O - Foot Patrol':
        folium.Marker(location=[crash["Latitude"], crash["Longitude"]],
            icon=folium.Icon(color='blue')).add_to(map_osm)
map_osm 
```

Am I able to visually interpret the data?

Yes, I am able to visually interpret the data and at the same time, not making the markers crowded by selecting certain arrest types as this is important to understand where these arrest occured. Based on the markers, I did BLUE as Foot Patrol arrests and RED as Motorcycle arrests.
Also, by selecting the types, not only I am able to prevent overcrowding the maps which means that I could also see where these arrest mostly occured. As we can see here, the 'L - Motorcycle' arrest are likely to happen around Aspen Hill and Rockville.

Why am I showing using this data and map?

I am showing this mainly because the map is the best way to see and visualize where these violations and arrests occured within the Montgomery County and also, with this, I could see where the police are mostly active because we know that not all police are active (or sometimes they just let these violations by as long as there are no bad consequences). I chose these arrests for my markers because we know that police are usually would be on their motorcycles or on foot walking around the neighborhood. Also, we can also see that most of the outliers are the BLUE marker which is the Foot Patrol arrests. Not only that, we can see that there is a cluster of ONLY BLUE happening around Veirs Mill Road.
